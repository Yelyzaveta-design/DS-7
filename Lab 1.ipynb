Here is the sample mean:

$$
\bar{y} = \frac{1}{n}{(\sum_{i=1}^n)y_i}
$$

Here is the sample variance:

$$
\sigma^2 = \frac{(\sum_{i=1}^n)(y_i - \bar{y})^2 }{n - 1}
$$

Here is the sample standart deviation: 

$$
\sigma = \frac{\sqrt{\sum_{i=1}^n (y_i - \bar{y})^2}}{n-1}
$$
