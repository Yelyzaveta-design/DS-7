## Дано матрицю. Знайти ранг матриці:

```python
a = np.array([[ 1,  4, 5],
              [7, 3, 2],
              [9, 16, 1]])
rank = np.linalg.matrix_rank(a)
print ('Ранг матриці a: ', rank)
```

## Знайти визначник матриці:

```python
print('Визначник матриці а: ', np.linalg.det(a))
```

## Знайти власні значення та власні вектори матриці: 

```python
own = np.linalg.eigh (a)
print (' Власні значення та власні вектори матриці a: ')
print (own)
```

## Розв'яжіть систему рівнянь:
$$2x + 5y = 1 $$

$$x - 10y = 3  $$
   

```python
M1 = np.array([[2, 5], [1, -10]])
v1 = np.array([1, 3])
results = np.linalg.solve(M1, v1)
print ('Відповідь: ', results)
```
