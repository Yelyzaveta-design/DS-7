Напишіть функції, що:

Послідовність чисел ганеруємо рандомно

```python
import random


#генеруэмо тисячу чисел в діапазоні від 0 до 10
list_with_random_numbers = [random.randint(0,100) for i in range(1000)]


list_min = list_with_random_numbers[100]
for x in list_with_random_numbers:
    if x < list_min:
        list_min = x
        print('Мінімальне значення: ', list_min)                    
```

```python
import random

list_with_random_numbers = [random.randint(0,100) for i in range(1000)]


list_max = list_with_random_numbers[0]
for i in list_with_random_numbers:
    if i > list_max:
        list_max = i
        print ('Максимальне значення: ', list_max)
```

```python
import random
list_with_random_numbers = [random.randint(0,100) for i in range(1000)]
a = 0
for i in list_with_random_numbers:
    a += i
    mean = a / float(len(list_with_random_numbers))
    print ('Середнє значення: ', mean)
```

```python
import random
list_with_random_numbers = [random.randint(0,100) for i in range(1000)]
if len(list_with_random_numbers) % 2 == 0:
    left_index = int(len(list_with_random_numbers) / 2) - 1 
    right_index = int(len(list_with_random_numbers) / 2)
    m = (list_with_random_numbers[left_index] + list_with_random_numbers[right_index]) / 2
else:
    inx = len(list_with_random_numbers) // 2
    m = list_with_random_numbers[inx]
print ('Медіана: ', m)
```

```python
import random

from math import sqrt
 
partial_sum = 0
partial_sum_squares = 0
x_i = list_with_random_numbers = [random.randint(0,100) for i in range(1000)]
n = 0
while x_i != 0:
    n += 1
    partial_sum += x_i
    partial_sum_squares += x_i ** 2
    x_i = int(input())
print(sqrt((partial_sum_squares - partial_sum ** 2 / n) / (n - 1)))

from math import sqrt

def stddev(lst):
    mean = float(sum(lst)) / len(lst)
    return sqrt(float(reduce(lambda x, y: x + y, map(lambda x: (x - mean) ** 2, lst))) / len(lst))
print ('Середнє квадратичне відхилення: ',stddev)
```
