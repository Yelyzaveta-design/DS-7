## Розв'язок

```python
fig, ax = plt.subplots(3, 1, figsize=(15, 8))

ax[0].plot(df['a'], label = 'Line A')
ax[0].legend(loc = 'lower right')
ax[1].plot(df['b'], label = 'Line B')
ax[1].legend(loc = 'center left')
ax[2].plot(df['c'])
```

## Зразок

Створити візуалізацію, аналогічно рисунку 
- використайте томатний колір та відстань між стовбцями 0.5

## Розв'язок

```python
fig, ax = plt.subplots(figsize=(15,8))
ax.bar(x = df.index, height = df.a, color='tomato', width=0.8, alpha=0.5)
ax.legend(['a'], loc='upper right')
```

## Зразок

Створити візуалізацію, аналогічно рисунку

## Розв'язок

```python
fig, ax = plt.subplots(figsize=(20, 8))
ax.plot(df['a'], marker='^', linewidth=2, markersize=7)
plt.grid(axis='x', alpha=1, linewidth=1)
ax.set_title('This is the Title', fontsize=12)
ax.set_ylabel('This is the Y-Axis', loc='center')
ax.set_xlabel('This is the X Axis', loc='center', fontsize = 22)
```
